```python
# to check whether a given number is prime or not
def is_prime(x):
    count=0
    for num in range(1,x//2+1):
        if x%num==0:
            count+=1
    if count==1:
        return True
    else:
        return False
    
    
is_prime(int(input("Enter the number:")))
```

```python
# write a python program to print the prime no.s present within a range
def prime_range(lw,up):
    for num in range(lw,up+1):
        if is_prime(num): # used this function to be applied in a given range
            print(num,end=" ")
            
prime_range(100,1000)
```

```python
# print all the perfect numbers present in a range using functions
def is_perfect(n):
    s=0
    for num in range(1,n//2+1):
        if n%num==0:
            s+=num
    if s==n:
        return True
    else:
        return False
    
is_perfect(28)
```

```python
f=int(input("first number:"))
s=int(input("second number:"))
while f<=s:
    if is_perfect(f):
        print(f,end=" ")
    f+=1
```
