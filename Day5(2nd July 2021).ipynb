```python
# whether a given number is perfect or not
# perfect:if sum of the factors of a given number=number itself
#6:1+2+3=6
#28:1,2,4,7,14=28
pr=int(input("Enter the number:"))
s=0
for num in range(1,pr//2+1):
    if pr%num==0:
        s+=num
if s==pr:
    print("perfect")
else:
    print("not perfect")
```

####  Nested loops
- loop that is executed within another loop
- loop in loop
    - inner loop is executed w r t outer loop

```python
# 1 to 5 
for num in range(1,6):
    print("Andhra Loyola")
```

```python
# 1 to 5 
for row in range(1,6): # 5 values: 1 2 3 4 5
    for col in range(1,4): # 3 values:1 2 3 for each iteration
        print(col,end=" ")
    print()
```

```python
for row in range(1,6): # 5 times
    for col in range(1,row+1): # col:1 ,1 2, 1 2 3
        print(col,end=' ')
    print()
```

```python
# prime numbers present in a given range (1,100):25
lw,up=int(input()),int(input())
primes=0
for num in range(lw,up+1):
    count=0
    for each in range(1,num//2+1):
        if num%each==0:
            count+=1
    if count is 1:
        print(num,end=" ")
        primes+=1
print()
print("No.of primes:",primes)
    
```

```python
f,s=int(input()),int(input())
while f<=s:
    count=0
    num=1
    while num<=f: # 1 to f
        if f%num==0:
            count+=1
        num+=1
    if count==2:
        print(f,end=" ")
    f+=1
```

```python
# print the perfect present in the given range using while loop
num=1
while num<=1000:
    s=0
    for val in range(1,num):
        if num%val==0:
            s+=val
    if s==num:
        print(num,end=" ")
    num+=1
```
