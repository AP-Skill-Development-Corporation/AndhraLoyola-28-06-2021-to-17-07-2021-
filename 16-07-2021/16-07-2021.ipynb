```python
# basic linear graph:
import matplotlib.pyplot as plt
x=[25,26,27,28,29,30]
y=[20000,30000,40000,50000,60000,70000]
y1=[28000,32000,49000,50000,65000,89000]
plt.plot(x,y,label='x,y axis',color='y',marker='o',linestyle='--')
plt.plot(x,y1,label='x,y1 axis',color='c',marker='d',linestyle='-.')
plt.xlabel('ages')
plt.ylabel('salary')
plt.legend()
plt.title('Basic Linear Graph')
plt.tight_layout()
plt.grid(True)
plt.style.use('fivethirtyeight')
#print(plt.style.available)
plt.show()
```

```python
#subplot
x=np.linspace(1,10,100)
y=np.log(x)
plt.figure(1)
plt.subplot()
plt.title('x,y=log(x)')
plt.plot(x,y)

plt.figure(2)
plt.subplot()
plt.title('x,y=log(x)**2')
plt.plot(x,y**2)
plt.show()
```

```python
#subplot
x=np.linspace(1,10,100)

plt.figure(1)
plt.subplot(2,2,1)
plt.title('linear')
plt.plot(x,x)

plt.subplot(2,2,2)
plt.title('cubic')
plt.plot(x,x**3)

plt.figure(2)
plt.subplot(2,2,1)
plt.title('log')
plt.plot(x,np.log(x))

plt.subplot(2,2,2)
plt.title('exponential')
plt.plot(x,x**2)
plt.show()
```

```python
#bar graph
import matplotlib.pyplot as plt
x=[25,26,27,28,29,30]
y=[20000,30000,40000,50000,60000,70000]
plt.bar(x,y,label='x,y axis',color='y')
plt.show()
```

```python
#barh graph
import matplotlib.pyplot as plt
x=[25,26,27,28,29,30]
y=[20000,30000,40000,50000,60000,70000]
plt.barh(x,y,label='x,y axis',color='y')
plt.show()
```

```python
#pie chart
import matplotlib.pyplot as plt
slices=[25,26,27,28,29]
labels=['python','html','sql','java','css']
explode=[0,0,0,0.1,0]
plt.title('pie chart')
plt.pie(slices,labels=labels,explode=explode,wedgeprops={'edgecolor':'black'},startangle=90,shadow=True)
plt.show()
```

```python
#histogram chart
import matplotlib.pyplot as plt
x=[14,19,27,22,30,28]
y=[10,20,30,40,50,60]
plt.title('histogram')
age=29
color='r'
plt.axvline(age,color=color,label='age')
plt.hist(x,y,edgecolor='black')
plt.legend()
plt.show()
```
